```python
import cv2 
import os

import scipy as sp
import scipy.misc

import imreg_dft as ird
import matplotlib.pyplot as plt


#image = cv2.imread('ref.png')
#template = cv2.imread('align.png') 

basedir = os.path.join('..', 'examples')
# the TEMPLATE
im1_color =  cv2.imread('ref.png')
# the image to be transformed
im2_color = cv2.imread('output.jpg')

im0 = cv2.cvtColor(im1_color, cv2.COLOR_BGR2GRAY) 
im1 = cv2.cvtColor(im2_color, cv2.COLOR_BGR2GRAY) 


result = ird.similarity(im0, im1, numiter=3)

#result = ird.translation(im0, im1)
#tvec = result["tvec"].round(4)
#timg = ird.transform_img(im1, tvec=tvec)

# Maybe we don't want to show plots all the time

if os.environ.get("IMSHOW", "yes") == "yes":
    import matplotlib.pyplot as plt
    ird.imshow(im0, im1, result["timg"])
    plt.show()
    
cv2.imwrite("imreg_dft.jpg", result["timg"])
```
